```python
'''
匯入套件
'''
# 操作 browser 的 API
from selenium.webdriver.chrome.service import Service
from selenium import webdriver

# 處理逾時例外的工具
from selenium.common.exceptions import TimeoutException

# 面對動態網頁，等待某個元素出現的工具，通常與 exptected_conditions 搭配
from selenium.webdriver.support.ui import WebDriverWait

# 搭配 WebDriverWait 使用，對元素狀態的一種期待條件，若條件發生，則等待結束，往下一行執行
from selenium.webdriver.support import expected_conditions as EC

# 期待元素出現要透過什麼方式指定，通常與 EC、WebDriverWait 一起使用
from selenium.webdriver.common.by import By

# 強制等待 (執行期間休息一下)
from time import sleep

# send_keys模擬鍵盤輸入
from selenium.webdriver.common.keys import Keys

# ChromeDriver 的下載管理工具
from webdriver_manager.chrome import ChromeDriverManager

# 下載檔案的工具
import wget

# 文字修飾
from pprint import pprint,pformat

'''
selenium 啓動 Chrome 的進階配置參數
參考網址：https://stackoverflow.max-everyday.com/2019/12/selenium-chrome-options/
'''
# 啟動瀏覽器工具的選項
my_options = webdriver.ChromeOptions()
# my_options.add_argument("--headless")                #不開啟實體瀏覽器背景執行
my_options.add_argument("--start-maximized")           #最大化視窗
my_options.add_argument("--incognito")                 #開啟無痕模式
my_options.add_argument("--disable-popup-blocking")    #禁用彈出攔截
my_options.add_argument("--disable-notifications")     #取消 chrome 推播通知
my_options.add_argument("--lang=zh-TW")                #設定為正體中文


# 使用 Chrome 的 WebDriver
# my_service = Service(executable_path="./chromedriver.exe")
driver = webdriver.Chrome(
    options = my_options,
#     service = my_service
)
```

```python

# 星數與評論數
# 使用等待機制等待至少一個符合條件的元素可見，最多等待10秒
elements = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.XPATH, '//span[@aria-label]'))
)
starandcomment = []
first_element_skipped = False  # 增加一個變數用來標記是否已經跳過第一個元素

# 遍歷所有元素，印出每個元素的 aria-label 屬性值，並跳過第一個元素
for element in elements:
    if not first_element_skipped:
        first_element_skipped = True
        continue  # 跳過第一個元素

    aria_label = element.get_attribute("aria-label")
    starandcomment.append(aria_label)
    print(aria_label)

# 打印所有星數與評論數
print(starandcomment)

print(len(starandcomment))

# number = len(starandcomment)
```

### 星數及標籤

```python
from selenium.common.exceptions import NoSuchElementException

five_start_and_commends = []
locationtags = []
hrefs = []

with open('小琉球景點.txt', 'r', encoding='utf-8') as file:
    href_read = eval(file.read())

# 匹配所有的 href
for sublist in href_read:
    href_pattern = sublist[1]
    hrefs.append(href_pattern)

for href in hrefs:
    try:
        driver.get(href)

        # 點擊按鈕
        btn = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[3]/div/div/button[2]')
        btn.click()

        # 抓星數多少評論
        elements = driver.find_elements(By.CSS_SELECTOR, 'tr.BHOKXe[aria-label]')
        merged_star = ""
        for element in elements:
            aria_label_value = element.get_attribute('aria-label')
            # 將每次循環的 aria_label_value 添加到 merged_string 中
            merged_star += aria_label_value + ', '

        # 去除最後的逗號和空格
        merged_string = merged_star.rstrip(', ')

        # 將合併後的字符串添加到列表
        five_start_and_commends.append(merged_string)

        # 檢查用打印最終的列表
        # print(five_start_and_commends)
        print(merged_star)
        sleep(2)

    except NoSuchElementException as e:
        print(f"Error in grabbing stars and comments: {e}")
        five_start_and_commends.append('沒有星數')

    try:
        tags = driver.find_elements(By.CSS_SELECTOR, 'span.uEubGf.fontBodyMedium')
        merged_tags = ""
        for tag in tags:
            merged_tag = tag.text
            # 將每次循環的 merged_tag 添加到 merged_tags 中
            merged_tags += merged_tag + ', '
        locationtags.append(merged_tags)
        print(merged_tags)
        sleep(2)

    except NoSuchElementException as e:
        print(f"Error in grabbing tags: {e}")
        locationtags.append('沒有標籤')

print(five_start_and_commends)
print(locationtags)
```
